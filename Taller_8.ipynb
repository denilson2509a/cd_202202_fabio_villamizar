```python
import pandas as pd
import statsmodels.api as sm
```

```python
df= pd.read_csv("https://raw.githubusercontent.com/adiacla/bigdata/master/titanic.csv")
df
```

```python
df.isna().sum()
```

```python
df=df.drop("Cabin",axis=1)
df
```

```python
df=df.dropna()
df
```

```python
df.dtypes
```

```python
x=df.iloc[:,[5,6,7]]
y=df.Survived
logit_model=sm.Logit(y,x)
result=logit_model.fit()
print(result.summary())
```

```python
# 1 para hombres, 0 para mujeres
df = pd.get_dummies(df, columns = ["Sex"], drop_first = True)
```

```python
df
```

```python
x = df.iloc[:,[4,5,6,10]]
y = df.Survived
logit_model = sm.Logit(y,x)
result = logit_model.fit()
print(result.summary())
```

```python
hashEmbarked={
    1:'S', 
    2:'C',
    3:'Q'
}

df['EmbarkedNum'] = df['Embarked'].replace(
  # Reemplaza cada valor por el valor en la segunda lista en el mismo índice del match.
  hashEmbarked.values(), # Lista de Strings
  hashEmbarked.keys() # Lista de Números
)
```
